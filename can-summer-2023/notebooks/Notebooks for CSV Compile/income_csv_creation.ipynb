## Imports

```python
#Importing necessary dependencies
import pandas as pd
import geopandas as gpd
from shapely import wkt
```

```python
income = pd.read_csv('../../data/clean/income_demographics.csv')
```

### Percents: employed vs unemployed

```python
income_demos_copy = income.copy(deep=True)
income_demos_copy.set_index('geoid',inplace=True)

old_columns = ['civilian population in labor force 16 years and over: employed',
              'civilian population in labor force 16 years and over: unemployed',]

new_columns = ['Percent Population in Labor Force 16+: Employed',
              'Percent Population in Labor Force 16+: Unemployed']

for i in range(1,len(new_columns)+1):
    #print(i)
    income_demos_copy[new_columns[i-1]] = income_demos_copy[old_columns[i-1]]/income_demos_copy['civilian population in labor force 16 years and over']
```

### Percents: Poverty status

```python
old_columns3 = ['population age 18 to 64 for whom poverty status  is determined: living in poverty',
              'population age 18 to 64 for whom poverty status  is determined: at or above poverty level']

new_columns3 = ['Percent Ages 18-64: Living in Poverty',
              'Percent Ages 18-64: At or Above Poverty Level']

for i in range(1,len(new_columns3)+1):
    #print(i)
    income_demos_copy[new_columns3[i-1]] = income_demos_copy[old_columns3[i-1]]/income_demos_copy['population age 18 to 64 for whom poverty status  is determined']
```
