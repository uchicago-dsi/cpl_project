```python
import pandas as pd
```

```python
age = pd.read_csv('../final_data/age_demos (1).csv')
commute = pd.read_csv('../final_data/commute_demos.csv')
education = pd.read_csv('../final_data/education_demos (1).csv')
income = pd.read_csv('../final_data/income_demos.csv')
internet = pd.read_csv('../final_data/internet_demos.csv')
race = pd.read_csv('../final_data/race_demos (1).csv')
```

```python
first_merge = age.merge(commute,on='geoid') 
first_merge = first_merge.drop(columns=['qualifying name_y', 'geometry_y'], axis=1)
first_merge.rename(columns={'qualifying name_x': 'qualifying name', 'geometry_x': 'geometry'}, inplace=True)
first_merge

second_merge = first_merge.merge(education,on='geoid') #.merge(internet,on='geoid').merge(race,on='geoid')
second_merge = second_merge.drop(columns=['qualifying name_y'], axis=1)
second_merge.rename(columns={'qualifying name_x': 'qualifying name'}, inplace=True)
second_merge

third_merge = second_merge.merge(income, on='geoid')
third_merge = third_merge.drop(columns=['qualifying name_y', 'geometry_y'], axis=1)
third_merge.rename(columns={'qualifying name_x': 'qualifying name', 'geometry_x': 'geometry'}, inplace=True)
third_merge

fourth_merge = third_merge.merge(internet, on='geoid')
fourth_merge = fourth_merge.drop(columns=['qualifying name_y', 'geometry_y'], axis=1)
fourth_merge.rename(columns={'qualifying name_x': 'qualifying name', 'geometry_x': 'geometry'}, inplace=True)
fourth_merge

final_merge = fourth_merge.merge(race,on='geoid')
final_merge = final_merge.drop(columns=['qualifying name_y', 'geometry_y'], axis=1)
final_merge.rename(columns={'qualifying name_x': 'qualifying name', 'geometry_x': 'geometry'}, inplace=True)
final_merge.head(1)

#df.rename(columns={"firstname": "FIRSTNAME", "lastname": "LASTNAME"}, inplace=True)
#check_values = first_merge['qualifying name_x'] == first_merge['qualifying name_y']
#check_values
```

```python
census_demos = final_merge.copy(deep=True)
census_demos.set_index('geoid',inplace=True)
```

```python
census_demos.to_csv('../../data/final_data/census_demos.csv')
```
